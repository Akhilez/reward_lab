```python
import itertools
import math
from matplotlib import pyplot as plt
```

```python
def get_sin_oscillator(freq, amp=1, phase=0, sample_rate=44100):
    phase = phase * math.pi / 180  #  Degrees to radians
    increment = (2 * math.pi * freq) / sample_rate
    return (
        math.sin(phase + v) * amp
        for v in itertools.count(start=0, step=increment)
    )
```

```python
osc = get_sin_oscillator(freq=2, sample_rate=512)
plt.plot([next(osc) for i in range(512)])
```

```python
osc = get_sin_oscillator(freq=2, phase=30, sample_rate=512)
plt.plot([next(osc) for i in range(512)])
```

```python
degrees = [0, 30, 190, 350]
oscillators = [
    get_sin_oscillator(freq=2, phase=phase, sample_rate=512)
    for phase in degrees
]
for i, o in enumerate(oscillators):
    plt.plot([next(o) for i in range(512)], label=degrees[i])
plt.legend()
```

```python
class Oscillator:
    def __init__(self, freq=1, amp=1, phase=0, sample_rate=50):
        self.freq = freq
        self.amp = amp
        self.phase = phase
        self.sample_rate = sample_rate

        self._t = -1

    def __next__(self):
        self._t += 1
        # self.phase += 1 * math.pi / 180
        # self.freq += 0.01
        increment = self._t * (2 * math.pi * self.freq) / self.sample_rate
        return math.sin(self.phase + increment) * self.amp


o = Oscillator(freq=1, amp=1, phase=0, sample_rate=50)
plt.plot([next(o) for _ in range(50)])
```
